```python
heart = heart.dropna()
```

```python
heart.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
#Converting string values into numerical value for prediction
for col in heart.columns.tolist():
    le = LabelEncoder()
    heart[col] = le.fit_transform(heart[col])
    
```

```python
# Putting all the collumns in variable x except for output collumn 
x = heart.iloc[:,:13]
```

```python
# Displaying top 5 records
x.head()
```

```python
# Putting number of arteries blocked collumn in  variable y
y = heart.loc[:,'num']
```

```python
# Displaying top 5 records
y.head()
```

```python
#Model selection - splitting the data for training an testing with 25%(Best pracitce among data scientists) testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
# Displaying number of rows and collumns of input data
x_train.shape
```

```python
# Displaying number of rows and collumns of testing input data
x_test.shape
```

```python
# Displaying number of rows and collumns of testing output data
y_test.shape
```

```python
# importing svm algorithm from sckit-learn
from sklearn import svm
```

```python
# training the model
sv = svm.SVC(kernel='linear',gamma='auto',C=2) #kernel transform linearly inseperable data into linearly seperable data
```

```python
# fitting input and output into algorithm 
sv.fit(x_train,y_train)
```

```python
# prediction on testing data
y_predict = sv.predict(x_test)
```

```python
# importing classification report from scikit learn metrics
from sklearn.metrics import classification_report
```

```python
# Comparing testing and predicted data
cr_s = classification_report(y_test,y_predict)
```

```python
# displaying classification report for SVM Algorithm
print(cr_s)
```

```python
from sklearn.tree import DecisionTreeClassifier
# Create the model 
dtc=DecisionTreeClassifier(criterion='gini',random_state=0)
# train the model using fit 
dtc.fit(x_train,y_train)
# predict on x
y_pred=dtc.predict(x_test)
# Evaluate
from sklearn.metrics import classification_report,accuracy_score
cr_dt=classification_report(y_test,y_pred)
print(cr_dt)
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
from sklearn.metrics import classification_report
cr_lr=classification_report(y_test,y_pred)
print(cr_lr)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
rfc = RandomForestClassifier()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.datasets import make_classification
param_grid = {'n_estimators': [10, 50, 100, 200, 500]}

grid_search = GridSearchCV(rfc, param_grid=param_grid, cv=5)

X, y = make_classification(n_samples=1000, n_features=10, n_classes=2)

grid_search.fit(x_train, y_train)

print("Best n_estimators:", grid_search.best_params_['n_estimators'])
```

```python
rfc = RandomForestClassifier(n_estimators=10, random_state=42)
rfc.fit(x_train,y_train)
y_pred=rfc.predict(x_test)
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
from sklearn.model_selection import cross_val_score
scores = []
for n in range(10, 201, 10):
    rfc = RandomForestClassifier(n_estimators=n, random_state=42)
    cv_scores = cross_val_score(rfc, x_train, y_train, cv=10)
    mean_score = np.mean(cv_scores)
    scores.append(mean_score)
best_n = (np.argmax(scores) + 1) * 10
print("The best value of n_estimators is:", best_n)
rfc = RandomForestClassifier(n_estimators=130, random_state=42)
rfc.fit(x_train,y_train)
y_pred=rfc.predict(x_test)
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
from sklearn.metrics import classification_report,confusion_matrix,recall_score,precision_score,accuracy_score
y_pred.sum()
y_test.sum()
cr_rf = classification_report(y_pred,y_test)
print(cr_rf)
```
